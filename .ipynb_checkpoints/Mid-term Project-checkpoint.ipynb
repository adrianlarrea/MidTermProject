# Hypothesis Testing

Traffic law and road safety changed in the year 2021, so since I have gathered data ranging from 2015-2021, I'm expecting to see a decline in casualties, soft injured and severe injured in accidents ocurred in the area of Barcelona

My Hypothesis Testing is that the mean casualty rate from 2016-2020 will be lower than the mean casualty rate from 2021
And will do the same with soft and severe injured

For this I will split my data in two sets, to be able to calculate those means

### Splitting the data sets to calculate the mean

```python
New_data1= New_data[New_data.Year.isin([2021])]
```

```python
New_data1
```

```python
New_data2= New_data[New_data['Year'] != 2021]
```

```python
New_data2
```

#### Calculating the mean casualties

```python
print("mean casualties by year 2021 is",np.sum(New_data1['Casualties']))
print("mean casualties by year 2016-2020 is",np.sum(New_data2['Casualties'])/5)
print("mean casualties for total accidents 2021 is",np.mean(New_data1['Casualties']))
print("mean casualties for total accidents 2016-2020 is",np.mean(New_data2['Casualties']))
```

#### Calculating the severely injured

```python
print("mean severely injured by year 2021 is",np.sum(New_data1['Severe_injured']))
print("mean severely injured by year 2016-2020 is",np.sum(New_data2['Severe_injured'])/5)
print("mean severely for total accidents 2021 is",np.mean(New_data1['Severe_injured']))
print("mean severely for total accidents 2016-2020 is",np.mean(New_data2['Severe_injured']))
```

#### Calculating the softly injured

```python
print("mean softly injured by year 2021 is",np.sum(New_data1['Soft_injured']))
print("mean softly injured by year 2016-2020 is",np.sum(New_data2['Soft_injured'])/5)
print("mean softly injured for total accidents 2021 is",np.mean(New_data1['Soft_injured']))
print("mean softly injured for total accidents 2016-2020 is",np.mean(New_data2['Soft_injured']))
```

### Hypothesis testing

#### Checking test for Deceased

```python
Deceased2021= New_data1['Casualties']
Deceased2021=Deceased2021.astype(float)
```

```python
from scipy import stats
```

```python
stats_value, pvalue = stats.ttest_1samp(Deceased2021,0.002867254317265786)
print('Stat_Score =  ',stats_value)
print('Pvalue =  ',pvalue/2)
```

The one-sided ttest has a p-value well below .05, and the Stat_Score is negative, so we reject Null Hypothesis.
Therefore is likely that the mean number of deceased will have decreased in 2021 

#### Checking test for Heavily Injured

```python
Heavyinjured2021= New_data1['Severe_injured']
Heavyinjured2021=Heavyinjured2021.astype(float)
```

```python
stats_value, pvalue = stats.ttest_1samp(Heavyinjured2021,0.027623947308057804)
print('Stat_Score =  ',stats_value)
print('Pvalue =  ',pvalue/2)
```

The one-sided ttest has a p-value well below .16, and se we cannot reject Null Hypothesis

#### Checking test for Soft Injured

```python
Softinjured2021= New_data1['Soft_injured']
Softinjured2021= Softinjured2021.astype(float)
```

```python
stats_value, pvalue = stats.ttest_1samp(Softinjured2021,1.571566667758954)
print('Stat_Score =  ',stats_value)
print('Pvalue =  ',pvalue/2)
```

The one-sided ttest has a p-value well below .45, and se we cannot reject Null Hypothesis

## Conclusions

Our tests show that while we can anticipate a decrease in the number of deceased in traffic accidents, we cannot foresee the outcome of severely and softly injured

It is way more likely to have an accident in Eixample neighborhood, due to its shape, long straight streets were there is the possibility of driving faster, plus, it is located at the center of the city, and some of the busiest streets are there.

April and August are the months with the least accidents. The reason behind is that April has the longest vacation period in Easter of the year and August is tipically the month were most business close as well as being a school vacation month

In Spain there are 15,52M men with driving license and 11,68M according to the DGT (National Traffic Department), but the percentage of accidents in our sample is higher than the proportion just mentioned. This implies that it is more likely to have an accident if a man is driving than a woman, regardless of being just more men driving. Other reasons might be like men are more used to driving in a car with both genders, or men are more inclined to risk in their youth (driving faster)


An overwhelming majority of the cases were pedestrians are involved, seldom is their fault.

Motorcycles are involved in about half of the accidents.

Other variables that out of the scope of the dataset might have a great influence in de decrease of accidents in year 2021 compared to the previous period 2016-2020 such as:

In 2020 Barcelona was declared by law a low emission zone. Therefore old cars, and high emission cars (old diesel cars), which are more likely to suffer severe injuries or death in case of accidents, cannot drive freely anymore.

Plus, Covid-19 lockdown shaped new ways of working, having more workers working from home.

Additionally, vehicle fleet renewal brings more safety in-built technology, further reducing the possibility of having an accident.

So, to sum up, our model was able to predict up to a certain point the decrease in deaths for the period 2021 where law changed compared to the previous years,but there are also more driving factors that help explain this trend. Hence, in order to fully comprehend the impact of the legislation more variables need to be in the analysis.
